```python
import json
import os

d1 = "/home/yzhang6375/autoPerson/personabench_main/eval_data/eval_data_v1/synthetic_data/community_0/eval_info/eval_info_all.json"
d2 = "/home/yzhang6375/autoPerson/personabench_main/eval_data/eval_data_v1/synthetic_data/community_1/eval_info/eval_info_all.json"
c1 = "/home/yzhang6375/autoPerson/personabench_main/eval_data/eval_data_v1/synthetic_data/community_0/"
c2 = "/home/yzhang6375/autoPerson/personabench_main/eval_data/eval_data_v1/synthetic_data/community_1/"
def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

def build_rag_corpus(community_path):
    conversation_data_all = load_json(os.path.join(community_path, 'private_data', f"noise_0.0", "conversation_data_all.json"))
    user_ai_interaction_data_all = load_json(os.path.join(community_path, 'private_data', f"noise_0.0", "user_ai_interaction_data_all.json"))
    purchase_history_data_all = load_json(os.path.join(community_path, 'private_data', f"noise_0.0", "purchase_history_data_all.json"))
    conversation_data_all_lookup = {entry["Name"]: entry["Data"] for entry in conversation_data_all}
    user_ai_interaction_data_all_lookup = {entry["Name"]: entry["Data"] for entry in user_ai_interaction_data_all}
    purchase_history_data_all_lookup = {entry["Name"]: entry["Data"] for entry in purchase_history_data_all}
    return conversation_data_all_lookup, user_ai_interaction_data_all_lookup,purchase_history_data_all_lookup
```

```python
data1 = load_json(d1)
data2 = load_json(d2)
corpus1 = build_rag_corpus(c1)
corpus2 = build_rag_corpus(c2)
```

```python
for eval_info_dict in data1[:3]:
    name = eval_info_dict["Name"]
    eval_info = eval_info_dict["Eval_Info"]
    conversation_data = corpus1[0][name]
    user_ai_interaction_data = corpus1[1][name]
    purchase_history_data = corpus1[2][name]
    documents_dict = {"conversation_data": conversation_data, "user_ai_interaction_data": user_ai_interaction_data, "purchase_history_data": purchase_history_data}
    QA = eval_info["qa"]
    # print(len(QA))
    res = 0
    for item in conversation_data:
        res+=len(item['Conversations'])
    # print(res+len(user_ai_interaction_data)+len(purchase_history_data))
    print(res,len(user_ai_interaction_data),len(purchase_history_data))
```
